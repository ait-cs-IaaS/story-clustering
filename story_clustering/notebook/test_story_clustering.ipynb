```python
import json
f = open('../data/corpus_test.json','r')
corpus_dict = json.load(f)
f.close()
```

```python
news_items = list()
for e in corpus_dict['docs']:
    elem = corpus_dict['docs'][e]
    news_item = dict()
    news_item['id'] = elem['doc_id']
    news_item['url'] = elem['url']
    news_item['lang'] = elem['language']
    news_item['title'] = elem['title']
    news_item['text'] = elem['content']
    news_item['date'] = elem['publish_time']
    news_item['tags'] = elem['keywords']
    news_items.append(news_item)
```
